```python
a = int(input("Enter the first number:"))
operator = input("enter the operator (+, -, *, /, %):")
b = int(input("Enter the second number:"))

if operator == "+" :
    print('Result:',a+b)
elif operator == "-":
    print('Result:',a-b)
elif operator == "*":
    print('Result:',a*b)
elif operator == "/":
    if b == 0:
        print('Division by zero is not allowed')
    else:
        print('Result:',a/b)
elif operator == "%":
    print('Result:',a%b)
else:
    print("Invalid operator")
    
```
